```python
#transaction_df = pd.read_csv(os.path.join(path_to_data, transaction_file))
#investment_exp_df = pd.read_csv(os.path.join(path_to_data, investment_exp_file))
#investment_segment_df = pd.read_csv(os.path.join(path_to_data, investment_segment))
#aum_df = pd.read_csv(os.path.join(path_to_data, aum_file))
activity_df = pd.read_csv(os.path.join(path_to_data, activity_file))
#test_df = pd.read_csv(os.path.join(path_to_data,test_file))
```

```python
activity_df.describe()
```

```python
activity_df['Activity_Type'] = activity_df['Activity_Type'].astype(str)
```

```python
type_df = pd.get_dummies(activity_df[['Activity_Type','Activity_Count']] )

for column in type_df.columns:
    if column == 'Activity_Count':
        continue
    type_df[column]= type_df[column]*type_df['Activity_Count']
    
type_df.head()
```

```python
type_df= type_df.drop('Activity_Count', axis = 1)

activity_df = pd.concat([activity_df, type_df], axis=1)
activity_df.head()
```

```python
activity_df.columns
```

```python
grouped_advisor_activity_df = activity_df.groupby(['Unique_Advisor_Id','Month']).agg({'Activity_Type_1': 'sum','Activity_Type_11':'sum','Activity_Type_12':'sum', \
                                                                                   'Activity_Type_13': 'sum','Activity_Type_14':'sum','Activity_Type_16':'sum', \
                                                                                   'Activity_Type_17': 'sum','Activity_Type_18':'sum','Activity_Type_19':'sum', \
                                                                                   'Activity_Type_2': 'sum','Activity_Type_22':'sum','Activity_Type_23':'sum', \
                                                                                   'Activity_Type_25': 'sum','Activity_Type_26':'sum','Activity_Type_27':'sum', \
                                                                                   'Activity_Type_30': 'sum','Activity_Type_4':'sum','Activity_Type_5':'sum', \
                                                                                   'Activity_Type_6': 'sum','Activity_Type_7':'sum','Activity_Type_9':'sum'\
                                                                                   
                                                                            
                                                                           }).reset_index()
```
